```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
class MyLR:
    #This is custom class to implement Single Linear Regression Model
    #class constructor
    def __init__(self):
        self.m = None
        self.b = None
    #Method to train the linear regression model
    def fit(self, X_train, y_train):
        numerator = 0    #sum((x - x̄) (y - ȳ))
        denominator = 0  #sum((x - x̄)^2)
        #loop to calculate the numerator and denominator for the value of m according to OLS formula
        for i in range(X_train.shape[0]):
            numerator += (X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean())
            denominator += (X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean())
        
        #calculating m
        self.m = numerator / denominator
        
        #calculating b
        self.b = y_train.mean() - (self.m * X_train.mean())
        print(self.m)
        print(self.b)
        
    #Method to make predictions using our linear regression model object
    def predict(self, X_test):
        return ((self.m * X_test) + self.b)
```

```python
df = pd.read_csv("C:/Users/chand/Downloads/placement.csv")
df.head()
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
lr = MyLR()
#training the simple linear regression model
lr.fit(X_train, y_train)
```

```python
X_test[0]
```

```python
#getting prediction from our model
y_pred = lr.predict(X_test)
#checking performance metrics
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
