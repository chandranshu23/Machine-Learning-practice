```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('USDtoINR.csv')
df.head()
```

```python
df.tail()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df['date_index'] = df['Date'].copy()
df = df.set_index('date_index')
df
```

```python
df_resampled = df.resample('MS').first()

print("\nResampled Data (First of Every Month):")
df_resampled
```

```python
#Re-set the index to the 'Date'
df_final = df_resampled.drop(columns=['Date'])
df_final
```

```python
df = df_final.copy()

#visualize the data
plt.figure(figsize = (12,6))
sns.lineplot(data = df, x = df.index, y = df.INR)
plt.show()
```

```python
# Check further for seasonality and trend by doing decompaosition
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df.INR, period = 12)
fig = decomposition.plot()
plt.show()
```

```python
#Upward Trend and clear seasonality

#Check if the time seires is stationary
# Rolling mean and std dev for 12 months

df['rollMean'] = df.INR.rolling(window = 12).mean()
df['rollStd'] = df.INR.rolling(window = 12).std()

#plot original series with rolling mean and std
plt.figure(figsize = (12,6))
sns.lineplot(data = df, x= df.index, y = df.INR, label = 'Original')
sns.lineplot(data = df, x= df.index, y = df.rollMean, label = 'Rolling Mean')
sns.lineplot(data = df, x= df.index, y = df.rollStd, label = 'Rolling Std')
plt.title('Rolling Statistics')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

#Define the function for stationarity check (only p-value based)
def test_stationarity(df2, var):
    df2['rollMean'] = df2[var].rolling(window = 12).mean()
    df2['rollStd'] = df2[var].rolling(window = 12).std()

    adf_result = adfuller(df2[var])
    p_value = adf_result[1]

    print(f'ADF p_value: {p_value:.4f}')
    if p_value < 0.05:
        print("The time series is stationary (reject H0)")
    else:
        print("The time series is not stationary (fail to reject H0)")

    #plot
    plt.figure(figsize = (12,6))
    sns.lineplot(data = df2, x= df2.index, y = df2[var], label = 'Original')
    sns.lineplot(data = df2, x= df2.index, y = df2.rollMean, label = 'Rolling Mean')
    sns.lineplot(data = df2, x= df2.index, y = df2.rollStd, label = 'Rolling Std')
    plt.title('Rolling Statistics')
    plt.show()

test_stationarity(df, 'INR')
```

```python
INR_df = df[['INR']].copy()
#try a shift diff of 1
INR_df['shift'] = INR_df.INR.shift(1)
INR_df['shiftDiff'] = INR_df['INR'] - INR_df['shift']
print(INR_df.head(20))
test_stationarity(INR_df.dropna(), 'shiftDiff')
```

```python
# now ARIMA
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plot_pacf(INR_df['shiftDiff'].dropna(), lags = 20)
plt.show()
```

```python
plot_acf(INR_df['shiftDiff'].dropna(), lags = 20)
plt.show()
```

```python
#Build ARIMA model
train = INR_df[:round(len(INR_df)*70/100)] # Take the first 70% data
print(train.tail()) #just to check where it ends

test = INR_df[round(len(INR_df)*70/100) :] # Take the last 30% data
print(train.head()) #just to check where it starts

model = ARIMA(train['INR'], order = (9,1,9)) # parameters: p,d,q
model_fit = model.fit()
pred = model_fit.predict(start = len(train), end = len(INR_df)-1)
INR_df['arimaPred'] = pred
print(INR_df.tail())
```

```python
#Plot
sns.lineplot(data = INR_df, x = INR_df.index, y = 'INR')
sns.lineplot(data = INR_df, x = INR_df.index, y = 'arimaPred')
plt.show()
#Conclusion: the ARIMA prediction is not good
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
#taking p as 11 because it must be less than the seasonal component which is 12 in this case, otherwise sarimax sees a redundant feature and raises error
model = SARIMAX(train['INR'], order = (1,1,1), seasonal_order = (1,1,1,12))
model_fit = model.fit()
prediction = model_fit.predict(start = test.index[0], end = test.index[-1])
INR_df['sarimaxPred'] = prediction
print(INR_df.tail())
```

```python
#Plot
sns.lineplot(data = INR_df, x = INR_df.index, y = 'INR')
sns.lineplot(data = INR_df, x = INR_df.index, y = 'arimaPred')
sns.lineplot(data = INR_df, x = INR_df.index, y = 'sarimaxPred')
plt.show()
#Conclusion: sarimax predictions are much more accurate
```

```python
futureDate = pd.DataFrame(pd.date_range(start = '2025-01-01', end = '2029-12-01', freq = 'MS'), columns = ['Dates'])
futureDate.set_index('Dates', inplace = True)
print(futureDate.head())

print(model_fit.predict(start = futureDate.index[0], end = futureDate.index[-1]))

#Plot
sns.lineplot(data = INR_df, x = INR_df.index, y = 'INR')
sns.lineplot(data = INR_df, x = INR_df.index, y = 'arimaPred')
sns.lineplot(data = INR_df, x = INR_df.index, y = 'sarimaxPred')
model_fit.predict(start = futureDate.index[0], end = futureDate.index[-1]).plot(color = 'black')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
INR_df = INR_df.dropna()
#arima evaluation
mae = mean_absolute_error(INR_df['INR'], INR_df['arimaPred'])
rmse = sqrt(mean_squared_error(INR_df['INR'], INR_df['arimaPred']))
r2 = r2_score(INR_df['INR'], INR_df['arimaPred'])
print(f'ARIMA MAE: {mae}, RMSE: {rmse}, r2: {r2}')
```

```python
#sarimax evaluation
mae = mean_absolute_error(INR_df['INR'], INR_df['sarimaxPred'])
rmse = sqrt(mean_squared_error(INR_df['INR'], INR_df['sarimaxPred']))
r2 = r2_score(INR_df['INR'], INR_df['sarimaxPred'])
print(f'ARIMA MAE: {mae}, RMSE: {rmse}, r2: {r2}')
```
