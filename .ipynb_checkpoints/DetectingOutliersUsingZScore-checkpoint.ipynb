```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/chand/Downloads/100-days-of-machine-learning-main/100-days-of-machine-learning-main/day42-outlier-removal-using-zscore/placement.csv")
df.sample(5)
```

```python
plt.figure(figsize = (16,5))
plt.subplot(1,2,1)
sns.kdeplot(df['cgpa'], fill = True)

plt.subplot(1,2,2)
sns.kdeplot(df['placement_exam_marks'], fill = True)

plt.show()
```

### Since cgpa is normally distributed we can apply Zscore based outlier detection on it, but not on Placement_exam_marks feature since it is skewed for it we will have to try another approach

```python
print("Mean value of cpga: ",df['cgpa'].mean())
print("std value of cpga: ",df['cgpa'].std())
print("Max value of cpga: ",df['cgpa'].max())
print("Min value of cpga: ",df['cgpa'].min())
```

```python
#Finding boundary values
print("Highest allowed ", df['cgpa'].mean() + 3*df['cgpa'].std())
print("Lowest allowed ", df['cgpa'].mean() - 3*df['cgpa'].std())
```

```python
#Finding the outliers
df[(df['cgpa'] > 8.80) | (df['cgpa'] < 5.1)]
```

# Trimming

```python
#Trimming
new_df = df[(df['cgpa'] > 5.1) & (df['cgpa'] < 8.81)]
new_df.sample(10)
```

```python
#Calculating the Zcore (same as using mean - 3std as boundary)
df['cgpa_zscore'] = (df['cgpa'] - df['cgpa'].mean()) / df['cgpa'].std()
df.head()
```

```python
df[df['cgpa_zscore'] > 3]
```

```python
df[df['cgpa_zscore'] < -3]
```

```python
new_df = df[(df['cgpa_zscore'] > -3) & (df['cgpa_zscore'] < 3)]
new_df
```

# Capping

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
df['cgpa'] = np.where(
    df['cgpa'] > upper_limit,
    upper_limit,
    np.where(
        df['cgpa'] < lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```
